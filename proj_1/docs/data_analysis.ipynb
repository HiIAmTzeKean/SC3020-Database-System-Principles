```python
import pandas as pd
```

```python
df = pd.read_csv('games.txt', sep='\t')
df.tail(5)
```

```python
df.describe()
```

```python
df = df.sort_values(by='FG_PCT_home', ascending=True)
# save sorted data to a new file
df.to_csv('games_sorted.txt', sep='\t', index=False)
```

```python
# NOTE 99 missing data rows
clean_df = df.dropna()
df[df.isna().any(axis=1)]
```

```python
pd.set_option('display.float_format', lambda x: '%.6f' % x)
def format_number(x):
    if isinstance(x, float):
        if x.is_integer():
            return f"{int(x)}"  # Convert whole floats to int
        else:
            return f"{x:.6f}".rstrip('0').rstrip('.')  # Trim trailing zeros
    return x

desc = clean_df.describe()
desc.applymap(format_number)
```

```python
# FG_PCT_home key is duplicated
clean_df[clean_df.duplicated(subset=["FG_PCT_home"], keep=False)].sort_values(by=["FG_PCT_home"])
```
